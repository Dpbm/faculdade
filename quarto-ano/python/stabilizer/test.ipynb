```python
import numpy as np
```

```python
Z = np.matrix([[1, 0], [0, -1]])
X = np.matrix([[0, 1], [1, 0]])
```

```python
state = np.matrix([[1], [0], [0], [0]])
```

```python
stab = np.kron(Z,Z)
```

```python
np.dot(stab, state)
```

```python
neg_Z = -1 * Z
```

```python
error_stab = np.kron(neg_Z, Z)
```

```python
np.dot(error_stab, state)
```

```python
another_state = np.matrix([[0], [1], [0], [0]])

# 01 --> Z-Z
```

```python
stab = np.kron(Z, neg_Z)
```

```python
np.dot(stab, another_state)
```

```python
stab_2 = np.kron(Z,Z)
```

```python
np.dot(stab_2, another_state)
```

```python
np.dot(another_state.T, stab_2)
```

```python
def expval(state, stab):
    return np.dot(np.dot(state.T, stab), state)
```

```python
expval(another_state, np.kron(Z,Z)), expval(another_state, np.kron(-Z,Z))
```
