```python
import random  # Importa a biblioteca para gera√ß√£o de n√∫meros aleat√≥rios

# Classe base Personagem que define os atributos e comportamentos comuns a todos os personagens
class Personagem:
    def __init__(self, nome, classe, vida, ataque_min, ataque_max):
        self.nome = nome  # Nome do personagem
        self.classe = classe  # Classe do personagem (Guerreiro, Mago, etc.)
        self.vida = vida  # Pontos de vida (HP)
        self.ataque_min = ataque_min  # Dano m√≠nimo de ataque
        self.ataque_max = ataque_max  # Dano m√°ximo de ataque
        self.arma = None  # Associa√ß√£o com a classe Arma (inicialmente sem arma)

    # M√©todo para equipar uma arma ao personagem
    def equipar_arma(self, arma):
        self.arma = arma
        print(f"{self.nome} equipou {arma.nome}!")

    # M√©todo para atacar outro personagem
    def atacar(self, outro_personagem):
        dano = random.randint(self.ataque_min, self.ataque_max)  # Gera um dano aleat√≥rio dentro do intervalo
        if self.arma:  # Se o personagem tiver uma arma equipada, adiciona dano extra
            dano += self.arma.dano_extra
        outro_personagem.vida -= dano  # Reduz a vida do oponente
        print(f"{self.nome} atacou {outro_personagem.nome} causando {dano} de dano!")

    # M√©todo para verificar se o personagem ainda est√° vivo
    def esta_vivo(self):
        return self.vida > 0

# Classe Guerreiro, herda de Personagem e define um ataque especial
class Guerreiro(Personagem):
    def __init__(self, nome):
        super().__init__(nome, "Guerreiro", 100, 15, 25)  # Define atributos espec√≠ficos para o Guerreiro

    # Habilidade especial do Guerreiro: Golpe Poderoso (causa 30% mais dano)
    def golpe_poderoso(self, outro_personagem):
        dano = int(random.randint(self.ataque_min, self.ataque_max) * 1.3)
        outro_personagem.vida -= dano
        print(f"{self.nome} usou Golpe Poderoso em {outro_personagem.nome} causando {dano} de dano!")

# Classe Mago, herda de Personagem e define um ataque especial
class Mago(Personagem):
    def __init__(self, nome):
        super().__init__(nome, "Mago", 80, 10, 30)  # Define atributos espec√≠ficos para o Mago

    # Habilidade especial do Mago: Bola de Fogo (sempre causa dano m√°ximo)
    def bola_de_fogo(self, outro_personagem):
        dano = self.ataque_max
        outro_personagem.vida -= dano
        print(f"{self.nome} lan√ßou Bola de Fogo em {outro_personagem.nome} causando {dano} de dano!")

# Classe Arqueiro, herda de Personagem e define um ataque especial
class Arqueiro(Personagem):
    def __init__(self, nome):
        super().__init__(nome, "Arqueiro", 90, 12, 22)  # Define atributos espec√≠ficos para o Arqueiro

    # Habilidade especial do Arqueiro: Flecha Dupla (dois ataques consecutivos)
    def flecha_dupla(self, outro_personagem):
        dano1 = random.randint(self.ataque_min, self.ataque_max)
        dano2 = random.randint(self.ataque_min, self.ataque_max)
        outro_personagem.vida -= (dano1 + dano2)
        print(f"{self.nome} disparou Flecha Dupla em {outro_personagem.nome} causando {dano1 + dano2} de dano!")

# Classe Arma, usada para adicionar dano extra a um personagem
class Arma:
    def __init__(self, nome, dano_extra):
        self.nome = nome  # Nome da arma
        self.dano_extra = dano_extra  # Dano adicional fornecido pela arma

# Criando personagens
guerreiro = Guerreiro("Arthur")
mago = Mago("Merlin")

# Criando e equipando uma arma para o guerreiro
espada = Arma("Espada M√°gica", 5)
guerreiro.equipar_arma(espada)

# Simulando a batalha
while guerreiro.esta_vivo() and mago.esta_vivo():
    guerreiro.atacar(mago)  # O guerreiro ataca primeiro
    if mago.esta_vivo():
        mago.bola_de_fogo(guerreiro)  # O mago ataca se ainda estiver vivo

# Verificando e anunciando o vencedor
vencedor = guerreiro if guerreiro.esta_vivo() else mago
print(f"\nüèÜ {vencedor.nome} venceu a batalha!")
```
