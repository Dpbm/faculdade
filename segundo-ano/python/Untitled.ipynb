```python
df = pd.read_csv("https://raw.githubusercontent.com/Vaibhav-Mehta-19/linear-regression-weather-dataset/master/weather.csv")
df.head()
```

```python
df.columns
```

```python
plt.scatter(df["MinTemp"], df["MaxTemp"])
```

```python
plt.scatter(df["MaxTemp"], df["Rainfall"])
```

```python
#https://towardsdatascience.com/introduction-to-machine-learning-algorithms-linear-regression-14c4e325882a

import numpy as np

x = df["MinTemp"].to_numpy().reshape(-1, 1)
y = df["MaxTemp"].to_numpy().reshape(-1, 1)

def linear_regression(x,y,epochs=200):
    n = len(x)
    a = np.zeros((n, 1))
    b = np.zeros((n, 1))
    
    alpha = 0.0001
    
    for _ in range(epochs):
        # calculate the linear function
        result = a * x+b
        # get the error for each value
        error = y - result
    
        # calculate new parameters a and b
        a = a - alpha * 2 * np.sum(error)/n 
        b = b - alpha * 2 * np.sum(error * x)/n

    a = a.tolist()[0][0] * -1
    b = b.tolist()[0][0] * -1
    return (a, b)
```

```python
a, b = linear_regression(x, y)
a,b
```

```python
y_points = []
for i in range(20):
    y_points.append(a * i + b)
y_points[:3]
```

```python
plt.scatter(df["MinTemp"], df["MaxTemp"])
plt.plot(range(len(y_points)), y_points, color="red")
plt.show()
```
